```python
from sympy import *
from sympy.plotting import plot
init_printing()
```

```python
x = Symbol('x')
f = x**3 - 50*x
g = -x**4 + 88*x**2 - 241
```

Графики функций в одной системе координат.

```python
a=plot(f,g)
```

График функции пересечения

```python
z = f-g
a = plot(z)
```

Координаты точек пересечения [x,y]

```python
x_array = solve(z)
[[x_el.n().round(2),f.subs(x, x_el).n().round(2)]for x_el in x_array]
```
